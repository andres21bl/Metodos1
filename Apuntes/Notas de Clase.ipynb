```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 2.
n = 5
type(x)
```

```python
func = np.zeros(n)
type (func)
```

```python
func
```

```python
func[0] = x
```

```python
def Update(f,x):
    
    f = (x/f + f)/2.
    return f
```

```python
Update(x,x)
```

```python
for i in range(1,n):
    func[i] = Update(func[i-1],x)
```

```python
func
```

```python
k = np.arange(0,n,1)
```

```python
plt.scatter(k,func)
plt.axhline(y=np.sqrt(2),color='r')
plt.yscale('log')
```
